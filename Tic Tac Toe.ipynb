```python
from IPython.display import clear_output

def display_board(board):
    
    print( " |" +  " |")
    print(board[1] + "|" + board[2] + "|" + board[3])
    print( " |" +  " |")
    print("-----")
    print( " |" +  " |")
    print(board[4] + "|" + board[5] + "|" + board[6])
    print( " |" +  " |")
    print("-----")
    print( " |" +  " |")
    print(board[7] + "|" + board[8] + "|" + board[9])
    print( " |" +  " |")
```

```python
def player_input():
    
    choice = ""
    while choice != "X" and choice != "O":
        choice = input("Please select your sign (X or O): ")
        
    if choice == "X":
        player1 = "X"
        player2 = "O"
    else:
        player1 = "O"
        player2 = "X"
    return (player1,player2)
```

```python
import random

def choose_first():
    if random.randint(0,1) == 0:
        start = "X"
    else:
        start = "O"
    return start
```

```python
def space_check(board, position):
    
    return board[position] == " "
```

```python
def full_board_check(board):
    
    
    for i in range(1,10):
        if space_check(board,i) == True:
            return False
    return True
```

```python
def player_choice(board):
    
    choice = 100
    
    while choice not in range(1,10) or not space_check(board,choice):
        choice = int(input("Please provide a number 1-9: "))
        if int(choice) not in range(1,10):
            print("Number must be in range 1-9")
        if space_check(board,choice) == False:
            print("Field already used. Select another one")

    return choice    
```

```python
def replay():
    
    again = "Placeholder"
    
    while again != "Y" and again != "N":
        again = input("Do you want to play again?(Y or N)?")
        
        if again != "Y" and again != "N":
            print("Please select either Y or N")
    if again == "Y":
        return True
    else:
        return False
        
```

```python
print('Welcome to Tic Tac Toe!')

while True:
    # Set the game up here
    board = [' ']*10
    player_marker1, player_marker2 = player_input()
    
    turn = choose_first()
    print(turn + ' will go first')
    
    play_game = input("Ready to play? Y or N? ")
    
    if play_game == "Y":
        game_on = True
    else:
        game_on = False
    while game_on:
        if turn == player_marker1:
        #Player 1 Turn
        
            display_board(board)
            
            position = player_choice(board)
            
            place_marker(board, player_marker1, position)
            
            if win_check(board, player_marker1):
                print("Player 1 won!")
                display_board(board)
                game_on = False
            else:
                if full_board_check(board):
                    print("No place left. End of the game")
                    display_board(board)
                    game_on = False
                else:
                    turn = player_marker2
        
        else:
#PLAYER 2
            display_board(board)
            
            position = player_choice(board)
            
            place_marker(board, player_marker2, position)
            
            if win_check(board, player_marker2) == True:
                print("Player 2 won!")
                display_board(board)
                game_on = False
            else:
                if full_board_check(board) == True:
                    print("No place left. End of the game")
                    display_board(board)
                    game_on = False
                else:
                    turn = player_marker1

    if not replay():
        break
```
